# FEMALES DIABETES DATABASE FROM INDIA

Dataset ini berasal dari National Institute of Diabetes and Digestive and Kidney Diseases. Tujuan dari dataset ini adalah untuk secara diagnostik memprediksi apakah seorang pasien memiliki diabetes, berdasarkan beberapa pengukuran diagnostik yang termasuk dalam dataset. Beberapa batasan diterapkan pada pemilihan data ini dari basis data yang lebih besar. Secara khusus, semua pasien di sini adalah wanita berusia minimal 21 tahun dengan keturunan Pima Indian.

Berikut variabel-variabel yang ada dalam diabetes dataset ini

1. Pregnancies (Kehamilan): Rasio - Karena jumlah kehamilan adalah bilangan bulat, memiliki jarak yang konsisten antar nilai, dan ada titik nol yang berarti (tidak pernah hamil).

2. Glucose (Glukosa): Rasio - Mengukur kadar glukosa dalam darah, yang memiliki jarak yang konsisten dan titik nol yang berarti (tidak ada glukosa).

3. BloodPressure (Tekanan Darah): Rasio - Karena tekanan darah diukur dalam satuan yang memiliki nol yang sebenarnya dan jarak yang konsisten.

4. SkinThickness (Ketebalan Kulit): Rasio - Mengukur ketebalan kulit dalam milimeter, memiliki titik nol dan jarak konsisten.

5. Insulin: Rasio - Kadar insulin dalam darah, memiliki nilai nol yang berarti dan jarak konsisten.

6. BMI (Indeks Massa Tubuh): Rasio - Menghitung rasio antara berat badan dan tinggi badan kuadrat, dengan titik nol dan jarak konsisten.

7. DiabetesPedigreeFunction: Rasio - Sebuah skor numerik yang mewakili kemungkinan diabetes berdasarkan sejarah keluarga. Meskipun secara teknis ini adalah hasil dari fungsi dan bisa dianggap sebagai rasio karena memiliki jarak yang konsisten dan titik nol.

8. Age (Usia): Rasio - Karena usia memiliki titik nol yang bermakna dan jarak antar data yang konsisten.

9. Outcome (Hasil): Nominal - Ini adalah kategori (apakah individu tersebut memiliki diabetes atau tidak), sehingga termasuk dalam skala nominal.

#### Diagram Polar

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print(data.head())

columns_to_use = ['Pregnancies', 'Glucose', 'BloodPressure', 'BMI']
data = data[columns_to_use]

data_mean = data.mean()

num_vars = len(columns_to_use)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

data_mean = data_mean.tolist()
data_mean += data_mean[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.plot(angles, data_mean, color='blue', linewidth=2, linestyle='solid')

ax.fill(angles, data_mean, color='blue', alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(columns_to_use)

plt.title('Average Values of Variables in Diabetes Dataset', size=16, color='blue', y=1.1)

plt.show()
```
